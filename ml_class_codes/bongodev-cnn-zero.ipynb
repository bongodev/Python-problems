# RGB vs Grayscale - Image Basics

```python
rgb_img = cv2.imread('/kaggle/input/bongodev-img/example-rgb.png')
rgb_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.title('Example RGB Image')
plt.show()
```

```python
print('rgb shape: ', rgb_img.shape)
rgb_img
```

```python
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.title('Example Grayscale Image')
plt.show()
```

```python
print('shape: ', gray_img.shape)
gray_img
```
