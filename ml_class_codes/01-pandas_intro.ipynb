### Some Pandas Theory

#### Building Blocks: Series and Dataframes

```python
# Series
ages = np.array([13, 25, 19])
series1 = pd.Series(ages)
print(series1)
```

```python
series2 = pd.Series(ages, index=['Sajon', 'Anjan', 'Shahjalal'])
print(series2)
```

```python
series2['Sajon']
```

```python
ages = np.array([13, 25, 19, 22])
series2 = pd.Series(ages, index=['Sajon', 'Anjan', 'Shahjalal'])
```

```python
dataf = pd.DataFrame([
    ['Sajon', 99, 'RUET'],
    ['Tousif', 5, 'CoU'],
    ['Sayedul', 5, 'CoU'],
    ],
  columns=['Name','Age', 'Campus'])
```

```python
df2 = pd.DataFrame({
    'A': 1.0,
    'B': pd.Timestamp('20240224'),
    'C': pd.Series(1, index=list(range(4)), dtype='float32'),
    'D': np.array([3, 4, 5, 6], dtype='int32'),
    'E': pd.Categorical(['train', 'test', 'train', 'test']),
    'F': 'foo'
})
```
